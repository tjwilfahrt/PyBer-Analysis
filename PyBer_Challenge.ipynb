### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "/Users/tyler/Desktop/Data_Bootcamp/Unit_2_Python/PyBer_Analysis/city_data.csv"
ride_data_to_load = "/Users/tyler/Desktop/Data_Bootcamp/Unit_2_Python/PyBer_Analysis/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides_city_type = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides_city_type
```

```python
# 2. Get the total drivers for each city type
total_drivers_city_type = city_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers_city_type
```

```python
#  3. Get the total amount of fares for each city type
total_fares_city_type = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares_city_type
```

```python
#  4. Get the average fare per ride for each city type. 
avg_fare_ride_city_type = pyber_data_df.groupby(["type"]).mean()["fare"]
avg_fare_ride_city_type
```

```python
# 5. Get the average fare per driver for each city type. 
avg_fare_driver_city_type = (total_fares_city_type / total_drivers_city_type)
avg_fare_driver_city_type
```

```python
#  6. Create a PyBer summary DataFrame.
pyber_summary_df = pd.DataFrame(
    {'Total Rides': total_rides_city_type,
        'Total Drivers': total_drivers_city_type,
        'Total Fares': total_fares_city_type,
        'Avg Fare Per Ride': avg_fare_ride_city_type,
        'Avg Fare Per Driver': avg_fare_driver_city_type})
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:.2f}".format)

pyber_summary_df["Avg Fare Per Ride"] = pyber_summary_df["Avg Fare Per Ride"].map("${:.2f}".format)

pyber_summary_df["Avg Fare Per Driver"] = pyber_summary_df["Avg Fare Per Driver"].map("${:.2f}".format)

pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df.head(10)
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares for each date where the indices 
# are the city type and date.
sum_of_fares = pyber_data_df.groupby(["type", "date"]).sum()[["fare"]]
sum_of_fares.head(10)
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
sum_of_fares = sum_of_fares.reset_index()
sum_of_fares.head(10)
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' to get the total 
# fares for each type of city by the date.
sum_of_fares_pivot = sum_of_fares.pivot(index = 'date', columns = 'type', values = "fare")
sum_of_fares_pivot
```

```python
# 5. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
sum_of_fares_pivot.index = pd.to_datetime(sum_of_fares_pivot.index)
```

```python
# 6. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
sum_of_fares_2 = sum_of_fares_pivot.loc[(sum_of_fares_pivot.index > '2019-01-01') & (sum_of_fares_pivot.index <= '2019-04-29')]
sum_of_fares_2
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
sum_of_fares_2.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
sum_of_fares_week = sum_of_fares_2.resample("W").sum()
sum_of_fares_week
```

```python
# 9. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
import matplotlib.pyplot as plt
# Use the graph style fivethirtyeight.
ax = sum_of_fares_week.plot(figsize = (20,5))
style.use('fivethirtyeight')
ax.set(title="Total Fare by City Type", 
      xlabel="Month",
      ylabel="Fare ($USD)")
plt.savefig("Analysis/PyBer_fare_summary.png")
plt.show()
```
